```python
#questao 8

numero1= float(input('digite um número'))
numero2= float(input('digite outro número'))
operacao = (input('Informe se vc quer somar, subtrair, multiplicar, dividir ou exponenciar'))

soma = numero1 + numero2
subtrair = numero1 - numero2
multiplicar = numero1 * numero2
dividir = numero1 / numero2
exponenciar = numero1 ** numero2

if(operacao == 'somar'):
    print('A soma dos dois números é:', soma)
    if (soma % 2 == 0):
     print ("O número é par")
    else:
     print ("O número é impar ")
    if(soma > 0 ):
      print('O número é positivo')
    else:
      print('O número é negativo')  
    if( soma // 1 == soma ):
      print('O número é inteiro')
    else:
      print('O número é decimal')    

elif(operacao == 'subtrair'):
    print('A subtração dos dois números é:', subtrair)
    if (subtrair % 2 == 0):
         print ("O número é par")
    else:
        print ("O número é impar ")
    if(subtrair > 0 ):
        print('O número é positivo')
    else:
        print('O número é negativo')  
    if( subtrair// 1 == subtrair ):
        print('O número é inteiro')
    else:
        print('O número é decimal')   

elif(operacao == 'multiplicar'):
    print('A multiplicação dos dois números é:', multiplicar)
    if (multiplicar % 2 == 0):
         print ("O número é par")
    else:
        print ("O número é impar ")
    if(multiplicar > 0 ):
        print('O número é positivo')
    else:
        print('O número é negativo')  
    if( multiplicar // 1 == multiplicar ):
        print('O número é inteiro')
    else:
        print('O número é decimal')   

elif(operacao == 'dividir'):
    print('A divisão dos dois números é:', dividir)
    if (dividir % 2 == 0):
         print ("O número é par")
    else:
        print ("O número é impar ")
    if(dividir > 0 ):
        print('O número é positivo')
    else:
        print('O número é negativo')  
    if( dividir // 1 == dividir ):
        print('O número é inteiro')
    else:
        print('O número é decimal')   

elif(operacao == 'exponenciar'):
    print('A exponeciação dos dois números é:', exponenciar)
    if (exponenciar % 2 == 0):
         print ("O número é par")
    else:
        print ("O número é impar ")
    if(exponenciar> 0 ):
        print('O número é positivo')
    else:
        print('O número é negativo')  
    if( exponenciar // 1 == exponenciar ):
        print('O número é inteiro')
    else:
        print('O número é decimal')
```

```python
#questao 12

numeros = []
soma = 0

for i in range(5):
    numero = float(input('Informe as 5 temperaturas: '))
    numeros.append(numero)
    soma += numero

media = soma / len(numeros)
menorValor = min(numeros)
maiorValor = max(numeros)

print("A média da temepratura é:", media)
print("O menor valor da temepratura é:", menorValor)
print("O maior da temepratura é:", maiorValor)
```
